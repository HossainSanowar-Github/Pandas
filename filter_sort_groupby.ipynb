```python
import pandas as pd
```

```python
data=pd.read_csv('train.csv')
df=data.copy()
```

# Filter

```python
#Rows where column is greater than 30
df[(df['Age']>30) & (df['Fare']>83)]
```

```python
df['Age']>30
```

# Sort

```python
#Sort values by col in asending order
df.sort_values('Age')
```

```python
#sort values by col in desending order
df.sort_values('Age',ascending=False)
```

```python
#`sort values by col1 in asending order, then col2 in desending order
df.sort_values(['Age','Fare'],ascending=[True,False])
```

# Groupby

```python
#Returns a groupby object values from one column
df.groupby('Sex')
```

```python
#Returns a groupby object values from multiple column
df.groupby(['Sex','Name'])
```

```python
#Return the mean value of col2 and grouped by the values of col1 
df.groupby('Sex')["Age"].mean()
```

```python
#create the pivot table that group by col1 and calculate the mean of col2, col3
df.pivot_table(index="Sex",values=["Fare","Age"])
```

```python
#Finds the average across all columns for every unique column 1 group 
import numpy as np
df.groupby('Sex').agg(np.mean)
```

```python
df1=df.drop(['Name','Sex','Cabin'],axis=1)
df2=df[['Age','Fare','Pclass']]
#Applies a function across each column
df2.apply(np.mean)
```

```python
#Applies a function across each row
df2.apply(np.max,axis=1)
```

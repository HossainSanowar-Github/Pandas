```python
import pandas as pd
```

```python
df1=pd.read_csv("ex1.csv")
df2=pd.read_csv("ex2.csv")
```

```python
#Add the rows in df1 to the end of df2(columns should be identical)
df=df1.append(df2)
```

```python
df
```

```python
#Add the column in df1 to the end of df2(row should be identical)
pd.concat([df1,df2],axis=1)
```

```python
#SQL style join; "col" have identical values "how" can be one of "left","right", "outer", "inner"
df1.join(df2,on='a',how="left")
```
